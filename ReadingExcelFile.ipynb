```python
import pandas as pd
```

```python
data_url = 'https://github.com/ine-rmotr-projects/project-files/files/4086772/playstore.xlsx'

playstore_df = pd.read_excel(data_url,
                             usecols=['App', 'Rating', 'Installs', 'Genres', 'Last_Updated'],
                             parse_dates=['Last_Updated'],
                             na_values=['nan'])
```
